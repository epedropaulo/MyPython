Também podemos gerar números aleatórios seguindo quase qualquer distribuição conhecida.

Para gerar, por exemplo, a matriz de correlação entre os dois vetores podemos usar o seguinte comando do Numpy:

Para garantir que sempre geremos os mesmos resultados aleatórios (muito útil para uma lista de exercícios!), nós vamos fixar uma semente, ou "seed", que garante que os números que brotarão serão sempre os mesmos.

```python
np.random.seed(2023)
np.random.normal(mu, sigma, 5) 
```

```python
mu, sigma = 0, 10
num_samples = 10
x = np.random.normal(mu, sigma, num_samples)
print(np.mean(x))
print(np.var(x))
print(np.sqrt(np.var(x)))
print(np.std(x))
```

## 2.3.2 Gráficos

Vamos investigar agora como gerar gráficos. Primeiro, vamos fazer um gráfico com número aleatórios. A ferramenta principal de gráficos para Python, matplotlib, é infelizmente um pouco complicada. Há alternativas; uma das mais interessantes é a plotnine, que se baseia no ggplot, a principal ferramente de gráficos em R.

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.random.normal(0, 1, 100)
y = np.random.normal(0, 1, 100)

plt.plot(x, y, 'bo')
plt.ylabel("this is the y-axis")
plt.xlabel("this is the x-axis")
plt.title("Plot of X vs Y")
# plt.savefig('Figure.pdf')  # Salva uma figura na pasta onde está o notebook
plt.show()  # Imprime a figura no próprio notebook
```

Para gerar gráficos, às vezes é conveniente conseguir criar um vetor com uma sequência de números. O np.arange(start, end+1, stepsize) faz isso, começando do primeiro elemento e indo até o segundo (excluindo este).

```python
x = np.arange(1, 10+1, 0.5) 
print(x) 
```

Para gerar uma sequência dados os dois extremos e um número de elementos no meio, usamos o np.linspace(start, end, num). Note que para definir pi precisamos primeiro importar a biblioteca math.

```python
import math
x = np.linspace(-math.pi, math.pi, num = 50)
print(x)
```

Agora, podemos gerar gráficos mais complicados. Por exemplo, podemos definir um grid de valores usando np.meshgrid(x, y) e olhar para o valor da função em cada elemento do grid. 

```python
import matplotlib.cm as cm
import matplotlib.mlab as mlab
y = x
X, Y = np.meshgrid(x,y)
```

```python
%whos
```

```python
f = np.cos(Y)/(1 + np.square(X))
contour_plot = plt.contour(X, Y, f)
plt.show()
```

```python
f_avg = (f - np.transpose(f))/2  # f.T for transpose or tranpose(f)
plt.imshow(f_avg, extent=(x[0], x[-1], y[0], y[-1])) 
plt.show()
```

```python
from mpl_toolkits.mplot3d import axes3d
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot_wireframe(X, Y, f_avg)
plt.show()
```

A biblioteca matplolib não é das mais intuitivas, mas há muitos recursos disponíveis na internet. É sempre o melhor (às vezes, único) lugar para aprender a fazer algum gráfico específico.

## 2.3.3 Indexando dados

Para selecionar partes de vetores ou matrizes, usamos os índices correspondentes. Note que Python indexa usando 0, ou seja, vetor[0] é o primeiro elemento do vetor.

Vamos começar criando uma matriz e considerar como escolher seus elementos.

```python
A = np.arange(1, 17, 1).reshape(4, 4).transpose()
print(A)
```

Para conseguir o elemento 15, precisamos da terceira coluna e quarta linha.

```python
print(A[2, 3])
```

Lembre que Python começa sempre com 0, ao contrário de R. Nesse caso, quem é o elemento A[1, 2]?

```python
print(A[1, 2])
```

Para selecionar mais de um elemento no vetor ou na matriz, podemos usar uma lista. Veja um exemplo abaixo usando os elementos [2, 3] da linha 0 e da linha 1.

```python
A[[[0],[1]], [2,3]]
```

Outra maneira de fazer isso é através da notação [begin:end], que especifica as entradas [begin, begin+1, begin+2,...,end-1]. A notação [begin:end:by] é parecida, mas o incremento é de "by" ao invés de 1.

```python
A[0:2, 2:4] 
```

```python
A[0:2, 0:4:2]
```

Para selecionar uma linha ou coluna inteira, basta indicar com :, que significa begin:(end+1).

```python
A[0:2,:]
```

```python
A[:, 2:4] 
```

```python
A[0,:]
```

O valor -1, quando usado como índice de um vetor, quer dizer o último elemento. É como se o - indicasse que o índice é contado a partir do final.

```python
v = [0, 1, 2, 3, 4]
v[-1]
```

```python
A[-1, -1] 
```

Em Numpy, também é possível indexar matrizes e vetores usando elementos booleanos (i.e., True or False).

```python
v = np.array([0, 1, 2, 3, 4])
v[[True, False, False, False, True]]
```

```python
ind = np.ones((4,), bool)
ind[[0,2]] = False
print(ind)
```

```python
A[ind, :]
```

Ou seja, note que acima indicamos primeiro as linhas e depois as colunas. Por default, se só um índice é dado ele atua nas linhas.

```python
A[ind]
```

Por último, podemos obter as dimensões de um vetor ou de uma matriz usando o atributo .shape.

```python
A.shape
```
