Começaremos falando sobre o significado de setores na bolsa.

Os setores da bolsa correspondem à classificação que a B3 dá a um conjunto de empresas que possuem similaridade quanto ao seu segmento de atuação no mercado.
Ou seja, a própria B3 faz uma classificação de ações que trabalham em áreas parecidas.

Para fazer o nosso gráfico de dispersão iremos pegar os setores de energia, e bens industriais.

ações do setor de energia: eletrobras (ELET3), Copel (CPLE6)

ações de setor de bens industriais: Azul (AZUL4), Gol Linhas Aéreas (GOLL4)

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
dados = pd.read_csv('b3.csv')
```

```python
# Ações de energia
copel = dados[dados['ticker'] == 'CPLE6']
copel_t = copel[5300:5500]  # Faixa temporal de 08/2019
vp_copel = copel_t['open']

eletrobras = dados[dados['ticker'] == 'ELET3']
eletrobras_t = eletrobras[5300:5500]
vp_eletrobras = eletrobras_t['open']
```

```python
# Ações de bens industriais
azul = dados[dados['ticker'] == 'AZUL4']
azul_t = azul[570:770]
vp_azul = azul_t['open']

gol = dados[dados['ticker'] == 'GOLL4']
gol_t = gol[3750:3950]
vp_gol = gol_t['open']
```

```python
# Série histórica do índice.
ibov = pd.read_csv('ibovespa.csv')
ibov_t = ibov[:200]
vp_ibov = ibov_t['Open']
```

```python
dados = pd.DataFrame({"Copel": list(vp_copel),
                    "Eletrobras": list(vp_eletrobras),
                    "Azul": list(vp_azul),
                    "Gol": list(vp_gol),
                    "Ibovespa": list(vp_ibov),})
```

```python
pd.plotting.scatter_matrix(dados,figsize=(13, 10), diagonal='kde', color='orange')

plt.tight_layout(pad=0.5)
plt.subplots_adjust(top=0.95)
plt.suptitle('Gráfico de Dispersão', fontsize=16)
plt.xticks(rotation=30)

plt.show()
```
